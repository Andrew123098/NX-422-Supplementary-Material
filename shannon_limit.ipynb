```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
sns.set_style(style="darkgrid")
```

```python
Q =  6 * 1000 * 1e-3
k = 1.75
y = 2 * np.log10(Q) - k
list_A = np.linspace(0.1, 5, 100)

ax = sns.lineplot(x=list_A, y=np.log10(list_A))
ax.set_xticks(np.arange(0, max(list_A)+0.5, 0.5))
ax.axhline(y=y, color='r', linestyle='--', label='Shannon limit')
ax.set_xlabel('GSA [mm^2]', fontsize=12)
ax.set_ylabel('log10(GSA)', fontsize=12)
ax.legend()
plt.tight_layout()
```
